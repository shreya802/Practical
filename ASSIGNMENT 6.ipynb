```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_csv = pd.read_csv("sales_data.csv", encoding='latin-1')
df_excel = pd.read_excel("sales_data.xlsx")
df_json = pd.read_json("sales_data.json")

df_csv.head()
df_csv.info
df_csv.isna().sum()
df_csv.isnull().sum()
df_csv.describe()
# df_csv.head()

# Dropping the unneccessary columns
df_csv = df_csv.drop(["ADDRESSLINE2", "TERRITORY"], axis = 1)

df_csv.fillna(0, inplace=True)
df_csv.drop_duplicates(inplace=True)

df_csv.head()

# Descriptive statistics

df_csv["ORDERNUMBER"].count()

# Calculating the sums of orders
df_csv['SALES'].sum()

# Calculating the average of orders
df_csv['SALES'].mean()

# Aggregating data by specific variables
product_catagory_sales = df_csv.groupby('COUNTRY')['SALES'].sum()
product_catagory_sales

sns.boxplot(x="STATUS", y = "MONTH_ID", data = df_csv)
plt.show()

sns.histplot(x= "MONTH_ID",data=df_csv)
plt.xlabel("Month ID")
plt.ylabel("Count")
plt.title("Distribution of Month")
plt.show()

sns.histplot(x="COUNTRY", data=df_csv)
plt.title("Country wise sales Distribution")
plt.show()


plt.figure(figsize=(20,9))
sns.histplot(x="COUNTRY", data=df_csv)
plt.title("Country wise sales Distribution")
plt.show()
```
